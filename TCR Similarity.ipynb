### Initialize an object for tcrdist3 method 

```python
mo = TCRSimilarityMeasurementFactory("tcrdist3")
```

### TCR similarity for one repertoire: default parameters
#### TCR similarity for alpha seqs

```python
result = mo.compute_distance(rep1=rep1, organism="mouse", seq_type="ALPHA")
result
```

#### TCR similarity for beta seqs

```python
result = mo.compute_distance(rep1=rep2, organism="mouse", seq_type="beta")
result
```

#### TCR similarity for alpha and beta seqs

```python
result = mo.compute_distance(rep1=rep2, organism="mouse", seq_type="alphabeta")
result
```

### Using another metrics
#### TCR similarity for alpha and beta seqs combined

```python
result = mo.compute_distance(rep1=rep1, organism="mouse", seq_type="both", default=False,
                                metric='nw_metric', open=1, extend=1)
result 
```

### TCR similarity for tow repertoires 

```python
result = mo.compute_distance(rep1=rep1, rep2=rep2, organism="mouse", seq_type="alphabeta", default=False,
                                metric='nb_vector_hamming_distance')
result 
```

### TCR similarity for one dataset
#### Passing the dataset as dataframe object 

```python
result = mo.compute_distance_from_dataset(df1=df1, organism="mouse", seq_type="ALPHA", default=False,
                                             metric='nw_hamming_metric', open=2, extend=2, source="dash")
result 
```

#### Passing the dataframe as a path

```python
result = mo.compute_distance_from_dataset(path1=path2, organism="mouse", seq_type="ALPHAbeta", source="dash")
result
```

### TCR similarity for tow datasets
#### Passing the datasets as dataframes

```python
result = mo.compute_distance_from_dataset(df1=df1, df2=df2, organism="mouse", seq_type="BOTH",
                                             default=False, metric='nb_vector_editdistance', source="dash")
result
```

#### Passing the datasets as paths

```python
result = mo.compute_distance_from_dataset(path1=path1, path2=path2, organism="mouse", seq_type="alphabeta",
                                             default=False, metric='nb_vector_editdistance', source="dash")
result
```

### Merging

```python
result1 = mo.compute_distance_from_dataset(df1=df1, organism="mouse", seq_type="BETA", source="dash")
result2 = mo.compute_distance_from_dataset(df1=df1, organism="mouse", seq_type="ALPHA", source="dash")
result1.merge_results(result2)
```

### Merging results with different indices 

```python
result1 = mo.compute_distance_from_dataset(df1=df1, organism="mouse", seq_type="ALPHA", default=False,
                                             metric='nw_hamming_metric', open=2, extend=2, source="dash")
result1
```

```python
result2 = mo.compute_distance_from_dataset(df1=df2, organism="mouse", seq_type="ALPHA", default=False,
                                             metric='nw_hamming_metric', open=2, extend=2, source="dash")
result2
```
